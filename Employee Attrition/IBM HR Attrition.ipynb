```python
hr_df.drop(columns = ['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], inplace = True)
```

```python
#outlier detection
def outliers_det(df,column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    up = q3 + 1.5*iqr
    lb = q1 - 1.5*iqr

    out = df[(df[column] > up) | (df[column] < lb)]
    return out
```

```python
selected_columns = hr_df.select_dtypes(include = ["number"])
for col in selected_columns:
    out = outliers_det(hr_df,col)
    if len(out)>0:
        print("Outliers found in",col,"and the count is",len(out))   
    else:
        print("No outliers found in", col)
```

```python
#outlier treatment
from scipy.stats import mstats
for col in selected_columns:
    hr_df[col] = mstats.winsorize(hr_df[col], limits=[0.05, 0.05])  #capping the 5th and 95th percentile values
```

```python
selected_columns = hr_df.select_dtypes(include = ["number"])
for col in selected_columns:
    out = outliers_det(hr_df,col)
    if len(out)>0:
        print("Outliers found in",col,"and the count is",len(out))   
    else:
        print("No outliers found in", col)
```

## Analyzing the Attrition distribution and key relationships.

```python
# Set plot style
sns.set(style="whitegrid")

# Analyze the distribution of Attrition
attrition_counts = hr_df['Attrition'].value_counts()

# Plot the distribution of Attrition
plt.figure(figsize=(6,4))
sns.barplot(x=attrition_counts.index, y=attrition_counts.values, palette='coolwarm')
plt.title('Attrition Distribution')
plt.ylabel('Number of Employees')
plt.xlabel('Attrition Status')
plt.show()

# Further exploration - Relation between Attrition and other key factors
plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='Age', data=hr_df, palette='coolwarm')
plt.title('Attrition vs Age')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=hr_df, palette='coolwarm')
plt.title('Attrition vs Monthly Income')
plt.show()
```

### 
Attrition Distribution:
A significant majority of employees have not left the company, while a smaller proportion have experienced attrition (i.e., left).

Attrition vs Age:
Employees who left the company tend to have a slightly younger median age compared to those who stayed. There is a visible concentration of younger employees among those who left.

Attrition vs Monthly Income:
Employees with higher Monthly Incomes seem less likely to leave. Those who left the company generally have lower incomes, which could indicate a correlation between salary and employee retention.

```python
from sklearn.preprocessing import LabelEncoder
# List of categorical features to be encoded using Label Encoding
cat_le = ['Attrition', 'Gender', 'MaritalStatus', 'OverTime']

# Iterate over each categorical feature to apply Label Encoding
for feature in cat_le:
    
    # Initialize the LabelEncoder
    label_encoder = LabelEncoder()
    
    # Fit the encoder on the feature and transform the feature values
    hr_df[feature] = label_encoder.fit_transform(hr_df[feature])
    
    # Print information about the encoding for each feature
    print(f"\n{feature}:")
    print(f"\tClasses  : {list(label_encoder.classes_)}")
    print(f"\tEncoding : {label_encoder.transform(label_encoder.classes_)}")
```
